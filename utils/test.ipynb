```python
import os.path

import requests
page_size = 100
user = "10515494"
base_url = f"https://api.zotero.org/users/{user}/items?format=bibtex&limit={page_size}"
headers = {"Zotero-API-Key": "IGllBMo1bbUwTZeswcCXNPUv", "Zotero-API-Version": "3"}

total_results = 1
page = 0
with open(os.path.join("library.bib"), 'w+', encoding="utf-8") as file:
    while page * page_size <= total_results:
        res = requests.get(f"{base_url}&start={page}", headers=headers)
        total_results = int(res.headers["Total-Results"])
        file.writelines(res.text + "\n")
        page += 1
```
